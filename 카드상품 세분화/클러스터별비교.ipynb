```python
data = data[~((data['연령'] == 20) & (data['Life_Stage'] == 7))]
data = data[~((data['연령'] == 20) & (data['Life_Stage'] == 6))]
data = data[~((data['연령'] == 20) & (data['Life_Stage'] == 5))]
data = data[~((data['연령'] == 30) & (data['Life_Stage'] == 5))]
data = data[~((data['연령'] == 30) & (data['Life_Stage'] == 6))]
data = data[~((data['연령'] == 30) & (data['Life_Stage'] == 7))]
data = data[~((data['연령'] == 40) & (data['Life_Stage'] == 1))]
data = data[~((data['연령'] == 40) & (data['Life_Stage'] == 7))]

data.shape
```

```python
data = data[~((data['연령'] == 50) & (data['Life_Stage'] == 1))]
data = data[~((data['연령'] == 60) & (data['Life_Stage'] == 1))]
data = data[~((data['연령'] == 60) & (data['Life_Stage'] == 2))]
data = data[~((data['연령'] == 60) & (data['Life_Stage'] == 3))]
data = data[~((data['연령'] == 70) & (data['Life_Stage'] == 2))]
data = data[~((data['연령'] == 70) & (data['Life_Stage'] == 4))]


data.shape
```

## 이상치정리

```python
cluster_1_data = data[data['Cluster'] == 1]
cluster_3_data = data[data['Cluster'] == 3]
cluster_0_data = data[data['Cluster'] == 0]
cluster_2_data = data[data['Cluster'] == 2]

print(cluster_0_data.shape)
print(cluster_1_data.shape)
print(cluster_2_data.shape)
print(cluster_3_data.shape)
```

## 분포확인

```python
numeric_columns = ['정규화_부유도', 'Life_Stage', '연령','회원여부_이용가능', '총이용금액', '이용금액_신판_B0M', '총유효카드수', '총연회비', '연체비율', '마케팅동의여부']
```

```python
for col in numeric_columns:

    # x축 범위 계산 (모든 클러스터의 데이터에서 최소/최대값을 통합적으로 설정)
    x_min = min(cluster_0_data[col].min(), cluster_2_data[col].min(), cluster_3_data[col].min())
    x_max = max(cluster_0_data[col].max(), cluster_2_data[col].max(), cluster_3_data[col].max())
    x_range = (x_min, x_max)

    # y축 범위 계산 (빈도수의 최대값 설정)
    y_max = max(
        cluster_0_data[col].value_counts(bins=30).max(),
        cluster_2_data[col].value_counts(bins=30).max(),
        cluster_3_data[col].value_counts(bins=30).max()
    )
    y_range = (0, y_max + 5000)  # 약간의 여백 추가


    plt.figure(figsize=(10,10))

    # Cluster 0
    plt.subplot(2,2, 1)
    plt.hist(cluster_0_data[col].dropna(), bins=30, alpha=0.7, color='blue', label='Cluster 0')
    plt.title(f'Cluster 0: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xlim(x_range)  # x축 범위 설정
    plt.ylim(y_range)  # y축 범위 설정
    plt.legend()

    # Cluster 1
    plt.subplot(2,2, 2)
    plt.hist(cluster_1_data[col].dropna(), bins=30, alpha=0.7, color='red', label='Cluster 1')
    plt.title(f'Cluster 1: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xlim(x_range)  # x축 범위 설정
    plt.ylim(y_range)  # y축 범위 설정
    plt.legend()

    # Cluster 2
    plt.subplot(2,2,3)
    plt.hist(cluster_2_data[col].dropna(), bins=30, alpha=0.7, color='orange', label='Cluster 2')
    plt.title(f'Cluster 2: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xlim(x_range)  # x축 범위 설정
    plt.ylim(y_range)  # y축 범위 설정
    plt.legend()

    # Cluster 3
    plt.subplot(2,2,4)
    plt.hist(cluster_3_data[col].dropna(), bins=30, alpha=0.7, color='green', label='Cluster 3')
    plt.title(f'Cluster 3: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xlim(x_range)  # x축 범위 설정
    plt.ylim(y_range)  # y축 범위 설정
    plt.legend()

    plt.tight_layout()
    plt.show()
```
