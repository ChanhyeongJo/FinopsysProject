```python
condition = (
    ((df_회원정보_12_selected['연령'] == 20) & (df_회원정보_12_selected['Life_Stage'].isin([5, 6, 7]))) |
    ((df_회원정보_12_selected['연령'] == 30) & (df_회원정보_12_selected['Life_Stage'].isin([1, 5, 6, 7]))) |
    ((df_회원정보_12_selected['연령'] == 40) & (df_회원정보_12_selected['Life_Stage'].isin([1, 7]))) |
    ((df_회원정보_12_selected['연령'] == 50) & (df_회원정보_12_selected['Life_Stage'] == 7)) |
    ((df_회원정보_12_selected['연령'] == 60) & (df_회원정보_12_selected['Life_Stage'].isin([1, 2, 3]))) |
    ((df_회원정보_12_selected['연령'] == 70) & (df_회원정보_12_selected['Life_Stage'].isin([2, 4])))
)

df_회원정보_12_selected_filtered = df_회원정보_12_selected[~condition]
```

```python
# VIP 등급별 데이터프레임과 등급명을 리스트에 저장
vip_dfs = [df_vip_07, df_vip_06, df_vip_05, df_vip_04, df_vip_NaN]
grades = ['07', '06', '05', '04', 'NaN']

# 각 데이터프레임에 대해 통계 요약을 출력
for df, grade in zip(vip_dfs, grades):
    if df.empty:
        print(f"\nVIP 등급 {grade} 데이터프레임이 비어있습니다.\n")
        continue  # 데이터프레임이 비어있을 경우 건너뜁니다

    print(f"\nVIP 등급 {grade} 통계 요약:")
    print(df.describe().to_string())  # .to_string()을 사용하여 전체 내용을 출력
    print("\n" + "="*50)  # 각 요약 사이에 구분선을 추가하여 출력의 가독성 향상
```

```python
for df, grade in zip([df_vip_07, df_vip_06, df_vip_05, df_vip_04, df_vip_NaN], ['07', '06', '05', '04', 'NaN']):
    if df.empty:
        print(f'VIP 등급 {grade} 데이터프레임이 비어 있습니다.')
        continue  # 빈 데이터프레임은 스킵
    
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x='거주시도명')
    plt.title(f'VIP 등급 {grade} - 거주시도명 분포')
    plt.xticks(rotation=90)
    plt.show()
```

```python
for df, grade in zip([df_vip_07, df_vip_06, df_vip_05, df_vip_04, df_vip_NaN], ['07', '06', '05', '04', 'NaN']):
    print(f"VIP 등급 {grade} - 회원여부_이용가능 비율:", df['회원여부_이용가능'].mean())
    print(f"VIP 등급 {grade} - 소지여부_신용 비율:", df['소지여부_신용'].mean())
    print("\n")
```

```python
for df, grade in zip([df_vip_07, df_vip_06, df_vip_05, df_vip_04, df_vip_NaN], ['07', '06', '05', '04', 'NaN']):
    if df.empty or 'Life_Stage' not in df.columns:
        print(f'VIP 등급 {grade} 데이터프레임이 비어있거나 Life_Stage 열이 없습니다.')
        continue  # 빈 데이터프레임이나 Life_Stage 열이 없는 경우 스킵

    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x='Life_Stage')
    plt.title(f'VIP 등급 {grade} - Life Stage 분포')
    plt.xticks(rotation=45)
    plt.show()
```
